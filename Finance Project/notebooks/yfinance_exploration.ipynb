```python
import pandas as pd
is_multi = isinstance(df2.columns, pd.MultiIndex)
is_multi
```

Tells us columns are in fact multi index

```python
# Create indexing based on ticker while leaving price as a column name
long_table = df2.stack(level = "Ticker")

print(long_table.head())
print(long_table.columns)
print(long_table.index.names)
```

```python
long = long_table.reset_index()
long
```

```python
# Format the dates properly, first check format
print(long["Date"].dtype)
```

```python
# Because im ocd lets still format it
long["Date"] = pd.to_datetime(long["Date"])
long = long.sort_values(["Ticker", "Date"]).reset_index(drop = True)
long
```

Data is clean, formatted and sorted.

```python
# CONFIRM TRANSFORM WORKS

long = df2.stack(level = "Ticker").reset_index()
long = long.rename(columns = {
    "Date":"date",
    "Ticker": "ticker",
    "Open" : "open",
    "High" : "high",
    "Low" : "low",
    "Close" : "close",
    "Volume" : "volume",
})
long = long.sort_values(["ticker", "date"]).reset_index(drop = True)

long.columns.name = None

long
```

```python
df1 = df1.rename(columns={
    "Price" : "price",
    "Date" : "date",
    "Close" : "close",
    "High" : "high",
    "Low" : "low",
    "Open" : "open",
    "Adj Close" : "adj_close",
    "Volume" : "volume"
})

df1["ticker"] = "AAPL"

cols = ['date', 'ticker', 'open','high', 'low', 'close', 'volume']

if "adj_close" in df1.columns:
    cols.insert(6, "adj_close")

df1.columns.name = None

df1 = df1.reset_index()

df1


```

```python
if hasattr(df1.columns, 'levels'):
    df1.columns = df1.columns.get_level_values(0)


df1.columns.name = None

df1 = df1.rename(columns = {"Date": "date"})

df1
```
